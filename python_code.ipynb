```python
import pandas as pd
df=pd.read_csv('orders.csv',na_values=['Not Available','unknown'])
df['Ship Mode'].unique()
```

```python
# df.rename(columns={'Order Id':'order_id','Ship Mode':'ship_mode'})
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.columns
# df.head(10)
```

```python
df['discount']=df['list_price']*df['discount_percent']*0.01   #make a discount column
```

```python
#find sale price and profit from given data
df['sale_price']=df['list_price']-df['discount']
df['profit']=df['sale_price']-df['cost_price']
```

```python
#change the datatype of date
#df.dtypes # object dtypes
df['order_date']=pd.to_datetime(df['order_date'],format='%Y-%m-%d')
# df.head(5)
```

```python
#df.dtypes
#drop the columns 
df.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
df.head(10)
```

```python
# !pip install sqlalchemy
# !pip install pyodbc
import sqlalchemy as sal
engine=sal.create_engine("mysql+pyodbc://mysql_dsn")
conn=engine.connect()
```

```python
# load the data into mysql using append query
df.to_sql('df_orders',con=conn,index=False,if_exists='append')
```
